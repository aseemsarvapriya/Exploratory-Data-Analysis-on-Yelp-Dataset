# 1: What are the top 10 rated restaurants in the city of Edwardsville in Illinois?

**We are looking to open a restaurant in the Edwardsville city of Illinois. First step would be to look for best rated restauarnts already there in the city.**


```python

%sql SELECT name, stars, state FROM business WHERE categories LIKE '%restaurant%' AND city == "Edwardsville" AND state = "IL" ORDER BY stars DESC LIMIT 10;
```

# 2: How many restaurants are there in the city?
**We would like to know how many restaurants are there in the city.**

```python
%sql SELECT COUNT(*) AS Total_Restaurants FROM business WHERE categories LIKE '%restaurant%' AND city == "Edwardsville" AND state = "IL";
```

# 3: What are the most popular types of restaurants in the city?
**This will give us the details of the famous cuisines that people prefer to eat in the city.**

```python
%sql SELECT categories, COUNT(*) FROM business WHERE categories LIKE '%restaurant%' AND city == "Edwardsville" AND state = "IL" GROUP BY categories ORDER BY COUNT(*) LIMIT 10;
```

# 4: What are the most popular dishes in the city's restaurants?
**This will give us the details of the popular dishes that people prefer to eat in the city.**

```python
%sql SELECT text, COUNT(*) FROM tip WHERE business_id IN (SELECT business_id FROM business WHERE categories LIKE '%restaurant%' AND city == "Edwardsville" AND state = "IL") GROUP BY text ORDER BY COUNT(*) LIMIT 10;
```

# 5: What is the average rating of restaurants in the city?
This will give us the average star ratings of all the restaurants in the city so we will have an idea of what kind of restaurants people prefer.

```python
%sql SELECT AVG(stars) FROM business WHERE categories LIKE '%Restaurants%' AND city == "Edwardsville" AND state = "IL";
```

# 6: What is the Total review count of the top 10 most reviewed restaurants in the city?
**With this, we could get to know the restaurants with most reviews.**

```python
%%sql SELECT r.business_id,b.name, COUNT(*) as review_count
    FROM review as r
    INNER JOIN business as b
    ON r.business_id = b.business_id
    WHERE b.city == "Edwardsville"
    GROUP BY r.business_id
    ORDER BY review_count DESC
    LIMIT 10
```

# 7: We will look into the bakeries that operates in Illinois
**We have decided to go for a bakery shop in the city. To have a better idea we would like to know how many bakeries are there in the state and their average ratings.**

```python
%%sql select state, city, round(avg(stars),2) as avg, sum(review_count) as Total_review, count(distinct business_id) as Num_of_Business, sum(review_count)/count(distinct business_id) as reviews_by_business
from business
where state='IL' and categories like '%Bakeries%'
GROUP BY city
ORDER BY reviews_by_business DESC
```

# 8: Average rating of Bakeries in the city of Edwardsville in Illinois.
**This will give us the average rating of the Bakeries in the city and the area they are located.**

```python
%%sql select postal_code, avg(stars) as Average_star
from business
where city='Edwardsville' and categories like '%Bakeries%'
group by postal_code
order by Average_star desc
limit 3;
```

# 9: Open bakeries or cafes in Edwardsville sorted by their zipcode



```python
%%sql select postal_code, count(*)
from business 
where (categories like '%Bakeries%' OR categories like '%cafe%') and postal_code IN ('62025', '62026','62048','62084','62087') 
GROUP BY postal_code
limit 10;
```

# 10: We will check the 'tip' table to see the customer texts in general for Bakeries  in Illinois state.
**Before we go into more bakery data in the city, We would also like to know how people in Illinois responds to the bakery stores by looking at their reviews text.**

```python
%%sql select t.user_id,t.business_id, t.text,b.name, b.review_count
from tip as t
inner join business as b 
on t.business_id=b.business_id
where b.categories like '%Bakeries%' and b.state='IL' and b.is_open=1
Group by t.business_id LIMIT 10
```

# 11: Identify attributes for bakeries in Edwardsville.

**This will give us info about the attributes that people of Edwardsville look for in a Bakery shop**

```python
%%sql select b.name, b.attributes, b.hours 
from business as b
where b.stars > (select AVG(stars) from business) and b.categories like '%Bakeries%' and b.city='Edwardsville' and b.is_open=1
group by b.name
order by b.stars DESC
LIMIT 5
```
